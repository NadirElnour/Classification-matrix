```python
df = pd.read_csv('oxides.csv')
df.head()
```

```python
data1 = df[['CaO','SiO2','Al2O3','Fe2O3','SO3','Rock Type']]
data1.head()
```

```python
sns.distplot(data1.CaO,hist=True, rug=False)
```

```python
df_corr = data1.corr()
df_corr
```

```python
sns.heatmap(df_corr)
```

```python
sns.jointplot(data=data1,x='CaO',y='SiO2',kind='hex')
```

```python
sns.jointplot(data=data1,x='CaO',y='SiO2', kind='reg')
```

```python
g = sns.PairGrid(data1,hue='Rock Type') # __init__(self, data, *, hue=None, hue_order=None, palette=None, hue_kws=None, vars=None, x_vars=None, y_vars=None, corner=False, diag_sharey=True, height=2.5, aspect=1, layout_pad=0.5, despine=True, dropna=False, size=None)
g.map_diag(sns.distplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
g = sns.PairGrid(data1,hue='Rock Type')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)
g.add_legend()
```

```python
x_vars = ["CaO", "SiO2", "Al2O3", "Fe2O3"]
y_vars = ["CaO"]
g = sns.PairGrid(data1, hue="Rock Type", x_vars=x_vars, y_vars=y_vars)
g.map_diag(sns.distplot, color=".3")
g.map_offdiag(sns.scatterplot)
g.add_legend()
```
